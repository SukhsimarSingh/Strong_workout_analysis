# Analysis Functions

## Exercise Selection

```python
exercise_name_list
```

```python
def exercise_dict_maker(exercise_name):
  exercise_list = []

  for i in range(len(df)):
    (exercise, weight) = df.iloc[i, 3], df.iloc[i, 5]
    date_time_list = df.iloc[i, 0].split()
    if exercise == exercise_name:
      exercise_list.append([date_time_list[0],weight])
  for i in range(len(exercise_list)):
    exercise_list[i][1] = round(exercise_list[i][1], 2)

  date_list = []
  for item in exercise_list:
    date_list.append(item[0])

  date_list = list(dict.fromkeys(date_list))

  exercise_dict = dict.fromkeys(date_list)

  for i in range(len(exercise_list)):
    if exercise_dict[exercise_list[i][0]] == None:
      exercise_dict[exercise_list[i][0]] = [exercise_list[i][1]]
    else:
      exercise_dict[exercise_list[i][0]].append(exercise_list[i][1])

  for i in exercise_dict.values():
    if len(i) < 7:
      i.extend([0]*(7-len(i)))

  return exercise_dict
```

```python
def set_list_maker(dictionary,n):
  set_list = []
  for i in dictionary.values():
    set_list.append(i[n])
  return set_list
```

```python
def exercise_df_maker(exercise_dictionary):
  set_list = []
  for i in range(7):
    set_list.append(set_list_maker(exercise_dictionary,i))
  exercise_df = pd.DataFrame(set_list)
  exercise_df = exercise_df.T
  exercise_df.columns = ['Set1','Set2','Set3','Set4','Set5','Set6','Set7']
  date_list = list(exercise_dictionary.keys())

  for i in range(len(date_list)):
    date_list[i] = date_list[i].replace('-','/')

  exercise_df['Dates'] = date_list

  exercise_df = exercise_df[['Dates', 'Set1', 'Set2', 'Set3', 'Set4', 'Set5', 'Set6', 'Set7']]

  exercise_df['Dates'] = pd.to_datetime(exercise_df['Dates'],format='%Y/%m/%d')

  return exercise_df
```

```python
def line_chart(dataframe, set, exercise_name):
  fig = px.line(dataframe, y="Set"+"{}".format(set), x="Dates",
                title= exercise_name)
  fig.show()
```
